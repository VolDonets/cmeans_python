```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
results = pd.read_csv('./data/results/iris_res.csv')
results.head()

dict_res_org = {'FSI': np.array(results['fsi']),
                'PC': np.array(results['pc']),
                'PEC': np.array(results['pec']),
                'PCES': np.array(results['pcaes'])}
dict_res_prop = {'Mah': np.array(results['mah']),
                'Mah_Inv': np.array(results['mah_inv']),
                'KLC': np.array(results['klc'])}

line_style = {'FSI': '-',
              'PC': '-x',
              'PEC': '-o',
              'PCES': '->',
              'Mah': '-<',
              'Mah_Inv': '-v',
              'KLC': '-^'}
```

```python
def norm(data_arr):
    min_val = np.min(data_arr)
    max_val = np.max(data_arr)
    data_arr = data_arr - min_val
    data_arr = data_arr / (max_val - min_val)
    return data_arr
```

# IRIS RESULTS

```python
for key in dict_res_org.keys():
    plt.plot(range(1, 21), norm(dict_res_org[key]), line_style[key], label=key)

plt.xlabel('Count Clusters')
plt.legend(loc='upper right')
plt.show()
```

```python
for key in dict_res_prop.keys():
    plt.plot(range(1, 21), norm(dict_res_prop[key]), line_style[key], label=key)

plt.xlabel('Count Clusters')
plt.legend(loc='upper right')
plt.show()
```
