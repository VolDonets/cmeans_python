```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
def plot_confusion_matrix(cm, class_names=None, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues, figname=None):
    """
    This function plots a confusion matrix.
  
    Args:
      cm: The confusion matrix to be plotted.
      class_names: Optional list of class names corresponding to the rows and columns of the confusion matrix.
      normalize: Whether to normalize the confusion matrix values (True) or keep them as counts (False).
      title: The title of the confusion matrix plot.
      cmap: The colormap to use for the confusion matrix plot.
  
    Returns:
      None
    """
    plt.figure(dpi=1000)

    # Normalize the confusion matrix if desired
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print("Confusion matrix without normalization")

    # Display the confusion matrix
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    # plt.imshow(cm, interpolation='bilinear', cmap=cmap)
    plt.colorbar()

    # Set the labels for the rows and columns if provided
    if class_names:
        tick_marks = np.arange(len(class_names))
        plt.xticks(tick_marks, class_names, rotation=45, ha='right')
        plt.yticks(tick_marks, class_names)
    else:
        plt.xticks(np.arange(len(cm[0])), np.arange(len(cm[0])))
        plt.yticks(np.arange(len(cm)), np.arange(len(cm)))

    # Annotate entries in the confusion matrix
    fmt = '.2f'
    # fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, format(cm[i, j], fmt),
                     ha="center", va="center",
                     fontsize=16,
                     color="white" if cm[i, j] > thresh else "black")
        pass

    # Set the title and show the plot
    if title:
        plt.title(title)
    plt.tight_layout()
    plt.grid(False)
    if figname:
        plt.savefig(figname, format='png')
    plt.show()
```

```python
# Example usage:
cm = np.array([
    [24, 0, 20],
    [3, 15, 10],
    [2, 0, 5],
])
class_names = ["Mature (Old)", "Immature (Old)", "Mature (Young)"]

# Using class names
plot_confusion_matrix(cm, class_names=class_names, title='Confusion Matrix with class names')
plot_confusion_matrix(cm, class_names=class_names, cmap=plt.cm.Oranges, title='Confusion Matrix with class names')
plot_confusion_matrix(cm, class_names=class_names, cmap=plt.cm.Greens, title='Confusion Matrix with class names')

# Without normalization and with different colormap
cm = np.array([
    [24, 0, 20],
    [3, 15, 10],
    [2, 0, 5],
])
plot_confusion_matrix(cm, cmap=plt.cm.Oranges, title='Confusion Matrix without normalization and Oranges colormap')
```
